```python
# Limit data to smaller sample (e.g., 500 per class)
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

# Define data_dir with the correct path to your ASL dataset
data_dir = '/content/asl_dataset/asl_alphabet_train/asl_alphabet_train'  # Update with your path

train_data = datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='training',
    shuffle=True,
    seed=42
)

val_data = datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation',
    shuffle=False,
    seed=42
)
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam

# Load MobileNetV2 base
base_model = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')

# Freeze base layers
base_model.trainable = False

# Add custom head
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
x = Dense(128, activation='relu')(x)
output = Dense(train_data.num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)

# Compile
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
# Train the model
EPOCHS = 20  # You can increase this if time permits

history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=EPOCHS
)
```

```python
import matplotlib.pyplot as plt

# Accuracy plot
plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'], label='Train Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Val Accuracy', marker='o')
for i, (train, val) in enumerate(zip(history.history['accuracy'], history.history['val_accuracy'])):
    plt.text(i, train + 0.01, f"{train:.2f}", ha='center', va='bottom', fontsize=8)
    plt.text(i, val + 0.01, f"{val:.2f}", ha='center', va='bottom', fontsize=8)
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Loss plot
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Train Loss', marker='o')
plt.plot(history.history['val_loss'], label='Val Loss', marker='o')
for i, (train, val) in enumerate(zip(history.history['loss'], history.history['val_loss'])):
    plt.text(i, train + 0.01, f"{train:.2f}", ha='center', va='bottom', fontsize=8)
    plt.text(i, val + 0.01, f"{val:.2f}", ha='center', va='bottom', fontsize=8)
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import cv2
import mediapipe as mp
from tensorflow.keras.models import load_model

# Load trained model
model = load_model("asl_model.h5")
class_names = list(train_data.class_indices.keys())

# Initialize MediaPipe
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Flip for selfie-view
    frame = cv2.flip(frame, 1)
    image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(image_rgb)

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            h, w, _ = frame.shape
            x_min = int(min([lm.x for lm in hand_landmarks.landmark]) * w)
            y_min = int(min([lm.y for lm in hand_landmarks.landmark]) * h)
            x_max = int(max([lm.x for lm in hand_landmarks.landmark]) * w)
            y_max = int(max([lm.y for lm in hand_landmarks.landmark]) * h)

            # Add margin and handle boundaries
            margin = 30
            x_min = max(0, x_min - margin)
            y_min = max(0, y_min - margin)
            x_max = min(w, x_max + margin)
            y_max = min(h, y_max + margin)

            hand_img = frame[y_min:y_max, x_min:x_max]

            # Ensure valid image
            if hand_img.shape[0] == 0 or hand_img.shape[1] == 0:
                continue

            hand_img = cv2.resize(hand_img, (224, 224))
            hand_img = np.expand_dims(hand_img, axis=0) / 255.0

            # Predict
            pred = model.predict(hand_img)
            predicted_class = class_names[np.argmax(pred)]

            # Show prediction on screen
            cv2.putText(frame, f"Sign: {predicted_class}", (x_min, y_min - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

            # Draw bounding box
            cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (255, 0, 0), 2)

    cv2.imshow("ASL Real-Time Recognition", frame)

    # Exit on 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
